## Create a dataset


```python
# Vars
obs_n = 200
a, b = 3.5, 8
fluctuations = 0.5
```

```python
# Features
X = 1 + 2 * np.random.random(obs_n)
# Targets
Y = b + a * X + fluctuations * np.random.randn(obs_n)
```

## Plot dataset to chart


```python
# Plot
plt.plot(X, Y, '+')
plt.show()
```
